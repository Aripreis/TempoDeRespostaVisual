```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
```

```python
file =  pd.read_csv('EXPERIMENTO1.csv', encoding = "ISO-8859-1")

df = file["Distancia em metros"]
listdf = df.values.tolist()
print(df)
visualTime = [0.192203]*100
```

```python
def tempoDeResposta(dist):
  vel = ((dist*2)/10)**(1/2)
  return vel
```

```python
velResp= []
for i in df:
  aux = tempoDeResposta((float(i.replace(',', '.'))))
  velResp.append(aux)

distM = []
for i in listdf:
  aux = float(i.replace(',', '.'))
  distM.append(aux)

print(velResp)
print(distM)
```

```python
media = [sum(velResp)/len(velResp)]*100
desvpad = np.std(velResp)
desvpadmedia = desvpad/(np.sqrt(100))
print('média: ',media[0])
print('desvio padrão:',desvpad)
print('desvio padrao da media: ',desvpadmedia)
```

```python
plt.xlabel('valores em segundos')
plt.ylabel('frequência')
plt.hist(velResp,bins=28,edgecolor='black')
```

```python
plt.figure()
plt.xlabel('testes')
plt.ylabel('valores em segundos')
plt.plot(velResp,'.-')
plt.plot(media,'-', label = 'média do experimento')#laranja
plt.plot(visualTime,'-',label='média da Revista Pesquisa Em Fisioterapia(2016)')#verde
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), shadow=True, ncol=3)

# Exibindo o gráfico
plt.tight_layout()
```
